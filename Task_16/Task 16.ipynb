##### Basic plot 

```python
import matplotlib.pyplot as plt
```

Create a Simple Line Plot

```python
x = [2, 4, 6, 8, 10]
y = [1, 3, 5, 7, 9]
```

```python
plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Line Plot')
plt.show()
```

Plot Multiple Lines on the Same Graph

```python
y2 = [1, 4, 6, 8, 10]
```

```python
plt.plot(x, y, label='odd Numbers')
plt.plot(x, y2, label='Even Numbers')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Multiple Line Plot')
plt.legend()
```

Add Markers to a Line Plot

```python
plt.plot(x, y, marker='o')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot with Markers')
plt.show()
```

Customize Line Styles and Colors

```python
plt.plot(x, y, linestyle='--', color='r', marker='o')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Customized Line Plot')
```

##### Scatter plot 

Create a Basic Scatter Plot

```python
plt.scatter(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Basic Scatter Plot')
```

Customize Marker Styles, Sizes, and Colors

```python
plt.scatter(x, y, s=100, c='g', marker='^')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Customized Scatter Plot')
```

Plot Multiple Scatter Plots on the Same Graph

```python
y3 = [2, 4, 6, 8, 10]
```

```python
plt.scatter(x, y, label='Dataset 1')
plt.scatter(x, y3, label='Dataset 2', marker='x')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Multiple Scatter Plots')
plt.legend()
```

Add Labels and Titles to the Scatter Plot

```python
plt.scatter(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot with Labels')
plt.legend(['Data Points'])
```

##### Bar Charts

Create a Vertical Bar Chart

```python
categories = ['O', 'B', 'h', 'D', 'F']
values = [5, 7, 3, 8, 4]
```

```python
plt.bar(categories, values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Vertical Bar Chart')
```

Create a Horizontal Bar Chart

```python
plt.barh(categories, values)
plt.xlabel('Values')
plt.ylabel('Categories')
plt.title('Horizontal Bar Chart')
```

Customize Bar Colors and Add Edge Colors

```python
plt.bar(categories, values, color='skyblue', edgecolor='black')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Customized Bar Chart')
```

##### Pie Charts

Create a Basic Pie Chart

```python
sizes = [15, 70, 45, 90]
labels = ['A', 'S', 'C', 'W']
```

```python
plt.pie(sizes, labels=labels)
plt.title('Basic Pie Chart')
```

Customize Slice Colors and Add a Legend

```python
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
plt.pie(sizes, labels=labels, colors=colors)
plt.title('Customized Pie Chart')
plt.legend(labels)
```

Explode a Slice of the Pie Chart

```python
explode = (0.1, 0, 0, 0)
plt.pie(sizes, labels=labels, colors=colors, explode=explode)
plt.title('Exploded Pie Chart')
plt.legend(labels)
plt.show()
```

Add Percentage Labels to Slices

```python
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%')
plt.title('Pie Chart with Percentages')
plt.legend(labels)
plt.show()
```
