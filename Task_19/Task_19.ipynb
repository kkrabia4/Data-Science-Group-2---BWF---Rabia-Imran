### Regression Plots

Linear Regression

```python
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset('iris')

sns.regplot(data=iris, x='sepal_length', y='petal_length')
plt.title('Linear Regression Plot of Sepal Length vs Petal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.show()
```

Multiple Linear regression plot 

```python

tips = sns.load_dataset('tips')


sns.lmplot(data=tips, x='total_bill', y='tip', hue='sex')
plt.title('Multiple Linear Regression Plot of Total Bill vs Tip by Sex')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
```

Residual Plot

```python

sns.residplot(data=tips, x='total_bill', y='tip')
plt.title('Residual Plot of Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Residuals')
plt.show()
```

### Distribution Plot

Histograms

```python

sns.histplot(data=iris, x='sepal_length', bins=10, kde=True)
plt.title('Histogram of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')

```

KDE Plot 

```python

sns.kdeplot(data=iris, x='sepal_length', hue='species', fill=True)
plt.title('KDE Plot of Sepal Length by Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Density')

```

Distribution Plot

```python
sns.histplot(data=iris, x='sepal_length', kde=True)
plt.title('Distribution Plot of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Density')

```

Box Plot

```python
sns.boxplot(data=iris, x='species', y='sepal_length')
plt.title('Box Plot of Sepal Length by Species')
plt.xlabel('Species')
plt.ylabel('Sepal Length (cm)')
plt.show()
```

Violin Plot 

```python
sns.violinplot(data=iris, x='species', y='sepal_length')
plt.title('Violin Plot of Sepal Length by Species')
plt.xlabel('Species')
plt.ylabel('Sepal Length (cm)')
plt.show()
```

Pair plot 

```python
sns.pairplot(iris, hue='species')
plt.suptitle('Pair Plot of Iris Dataset', y=1.02)
plt.show()
```

Heat map 

```python
corr = iris.drop('species', axis=1).corr()

sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Heatmap of Iris Correlation Matrix')
plt.show()
```

Joint Plot

```python
# Create a joint plot
sns.jointplot(data=iris, x='sepal_length', y='sepal_width', kind='scatter')
plt.suptitle('Joint Plot of Sepal Length and Sepal Width', y=1.02)
plt.show()
```

Facet grid

```python
# Create a facet grid
g = sns.FacetGrid(iris, col='species')
g.map(sns.scatterplot, 'sepal_length', 'sepal_width')
g.add_legend()
g.fig.suptitle('Facet Grid of Sepal Length vs Sepal Width by Species', y=1.02)
plt.show()
```
