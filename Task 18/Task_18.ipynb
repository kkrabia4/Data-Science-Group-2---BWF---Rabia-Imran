### Scatter Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset
iris = sns.load_dataset('iris')
```

```python
print(iris.describe())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('iris')


print(df.head())
```

```python
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species')
plt.title('Scatter Plot of Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()
```

### Line Plot

```python
sns.lineplot(data=iris, x='sepal_length', y='petal_length', hue='species')
plt.title('Line Plot of Sepal Length vs Petal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.show()
```

### Bar Plot

```python
sns.barplot(data=iris, x='species', y='sepal_length')
plt.title('Bar Plot of Sepal Length by Species')
plt.xlabel('Species')
plt.ylabel('Sepal Length (cm)')
plt.show()
```

### Count Plot

```python
sns.countplot(data=iris, x='species')
plt.title('Count Plot of Iris Species')
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()
```

### Box Plot

```python
sns.boxplot(data=iris, x='species', y='sepal_length')
plt.title('Box Plot of Sepal Length by Species')
plt.xlabel('Species')
plt.ylabel('Sepal Length (cm)')
plt.show()
```

### Customization

Setting style

```python
sns.set_style('whitegrid')
sns.boxplot(data=iris, x='species', y='sepal_width')
plt.title('Box Plot of Sepal Width by Species with Whitegrid Style')
plt.xlabel('Species')
plt.ylabel('Sepal Width (cm)')
plt.show()
```

Addind title

```python
sns.set_style('darkgrid')

sns.scatterplot(data=iris, x='sepal_length', y='petal_length', hue='species')
plt.title('Scatter Plot of Sepal Length vs Petal Length with Darkgrid Style')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.show()
```

Changing palette

```python
sns.set_palette('Set2')
sns.barplot(data=iris, x='species', y='petal_length')
plt.title('Bar Plot of Petal Length by Species with Set2 Palette')
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')
plt.show()
```
